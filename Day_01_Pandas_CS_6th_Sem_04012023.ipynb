# Notes Of ML

Machine Learning Steps :-

1. Data Gathering
2. Data Preparation
3. Data Wrangling (Data Clean karna)
4. Data Analysis
5. Training The Data
6. Testing The Data
7. Deploying The Model



1. Data Gathering :- The first step is to collect the data from different sources such as files , data bases , mobiles , internet , etc. the more is the data available , more efficient training can be given to a machine model .
 
A. identifying the data sources

B. collecting the data from these sources

C. integerating the data obtained from these sources

These are the steps involved in data gathering . by performing these steps we get a coverent set of data known as data set .



2. Data Preparation :- The Preparation involves two processes :-

A. Data exploration - In these we try to establish relationship b/w different data entries such as co-relations , outliers and general trends

B. Data preprocessing - This is used to understand the nature of the data such as quality , format , shape , size , find the values etc..



3. Data Wrangling :- this is step is used to clean the data and prepare the raw data into a usable format . the data collected can have various issues such as missing value , invalid data , duplicate data etc..

so, data wrangling is used to handle all these issues.



4. Data Analysis :- Data analysis involves selection of analytical techniques and building the models . it starts with the determination of problem finding and selecting the type of problem , whether it is classification , regeration , association , clustering , etc. then model is build using this data and is evaluated.



5. Training The Data :- Training is important so that machine can learn different patterns , features and rules to improve its performance for better prediction.



6. Testing The Data :- After training testing is performed so that the accuracy and confidence of the machine can be checked.

7. Deploying The Model :- Once the testing results are verified the machine is all set to be deployed in real world.
